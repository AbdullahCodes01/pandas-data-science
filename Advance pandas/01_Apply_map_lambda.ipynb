```python
import pandas as pd
import numpy as np
```

MAP() — Element wise replace / transform

```python
df = pd.DataFrame({
    "Name": ["Arman","Riya","Kunal"],
    "DeptCode": [1,2,1],
    "Salary": [98000,55000,67000]
})

dept_map = {
    1: "IT",
    2: "HR"
}

df["Department"] = df["DeptCode"].map(dept_map)
df
```

APPLY() — Custom logic (Power tool)

```python
# Column apply
df["Bonus"] = df["Salary"].apply(lambda x: x*0.10)
df
```

```python
# Function + Apply
def level(Salary):
    if Salary > 60000:
        return "High"
    else:
        return "Normal"

df["Level"] = df["Salary"].apply(level)

#another method using lambda
df["post"] = df["Salary"].apply(lambda x: "Senior" if x>60000  else "Junior")
```

```python
df
```

```python
# row wise apply
df["Tax"] = df.apply(lambda row : row["Salary"] * 0.05 if row["Department"] == "IT" else row["Salary"] * 0.03, axis=1)
df
```

```python
# LAMBDA — Short function
df["Salary"] = df["Salary"].apply(lambda x: x+5000)
```
