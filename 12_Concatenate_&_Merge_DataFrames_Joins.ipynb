12. Concatenate & Merge DataFrames (Joins)

```python
import pandas as pd

df1 = pd.DataFrame({
    "EmpID": [1, 2, 3],
    "Name": ["Aman", "Riya", "Kunal"]
})

df2 = pd.DataFrame({
    "EmpID": [4, 5],
    "Name": ["Sneha", "Vikram"]
})
```

```python
# CONCATENATE (pd.concat)

result = pd.concat([df1,df2])
print(result)
```

```python
# for Clean index 
print(pd.concat([df1,df2],ignore_index=True))
```

```python
# Concatenate COLUMN-wise
df3 = pd.DataFrame({
    "Salary": [50000,40000,35000]
})
result2 = pd.concat([df1,df3],axis=1)
print(result2)
```

```python
# When columns mismatch
df4 = pd.DataFrame({
    "EmpID": [6],
    "Name": ["Pooja"],
    "Department": ["IT"]
})

pd.concat([df1,df4],ignore_index=True)
```

```python
# MERGE (JOINS)
employees = pd.DataFrame({
    "EmpID": [1, 2, 3, 4],
    "Name": ["Aman", "Riya", "Kunal", "Sneha"]
})

salary = pd.DataFrame({
    "EmpID": [1, 2, 4, 5],
    "Salary": [50000, 48000, 70000, 60000]
})
```

```python
# pd.merge(left_df, right_df, on="CommonColumn", how="type"). syntax
```

```python
# NNER JOIN (default)
pd.merge(employees, salary, on="EmpID", how="inner")
```

```python
# LEFT JOIN
pd.merge(employees,salary,on="EmpID",how= "left")
```

```python
# RIGHT JOIN
pd.merge(employees,salary,on="EmpID",how="right")
```

```python
# OUTER JOIN
pd.merge(employees,salary,on="EmpID",how="outer")
```

```python
# Merge on DIFFERENT column names
dfA = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["A", "B", "C"]
})

dfB = pd.DataFrame({
    "EmpID": [1, 3],
    "Salary": [50000, 60000]
})

pd.merge(dfA,dfB,left_on="ID",right_on="EmpID")
```

```python
# Concatenate two DataFrames row-wise
pd.concat([df1,df2])
```
