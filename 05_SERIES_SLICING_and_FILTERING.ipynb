5. SLICING & FILTERING

there are 2 type of series indexing 

1.Position based (0,1,2...)

2.label-based ("Ali","Neha"...)

```python
import pandas as pd
import numpy as np
```

```python
# 1. Normal slicing (position based)

s = pd.Series([1,2,3,4,5])
print(s[0]) #first element
```

```python
s[1:4] # 1 to 4th element
```

```python
s[:3] #first three elements
```

```python
s[::2] #step by 2. (1,3,5)
```

```python
# 2. Label based slicing
s1 = pd.Series([60,70,80,90],index=["Arman","Abdullah","Sadiya","Farhan"])
s1["Farhan"] # single value
```

```python
s1[:"Sadiya"] # first to sadia
```

```python
s1["Arman": "Abdullah"] # arman to abdullah (include abdullah)
```

```python
# 3. Multiple selection
s1[["Arman","Sadiya"]]
```

```python
# 4. Filtering with condition (MOST IMPORTANT)
s1[s1>70] 
```

```python
s1[(s1>50)& (s1<90)]
```

```python
s1[s1==70]
```

```python
# 5. Using logical operators
(s1 > 80) | (s1 == 70)
```

```python
(s1 >= 80) & (s1 <= 90)
```

```python
# 6. Filtering with isin()
s1[s1.isin([70,80])]
```

```python
# 7. Filtering with notnull()
s1[s1.isnull()]
```

```python
mask = s1 > 70
print(mask)
print(s1[mask])
```

```python
# 9. Change values using filtering

s1[s1<70] = 90
```

```python
# Mini Practice â€“ Pandas Series
# Create a Series with 6 values.

salary = pd.Series([100000,45000,60000,55000,46000,34000],index=["abdullah","drishti","arushi","cristina","mehak","kamakshi"])
print(salary)
```

```python
# Print the values that are greater than the mean of the Series.

print(salary[salary > salary.mean()])
```

```python
# Replace all salary less than 50000 with 55000
salary[salary<50000] = 55000 
print(salary)
```

```python
# Print only two selected (named) indexes from the Series.
salary[["abdullah","drishti"]]
```

```python
# Print alternate values from the Series.
salary[::2]
```
