```python
import pandas as pd
import numpy as np

sales_df = pd.DataFrame({
    "OrderID": [1001,1002,1003,1004,1005,1006,1007,1008,1009,1010,1011,1012],
    "Customer": ["Arman","Riya","Kunal","Sneha","Vikram","Pooja","Rahul","Neha","Arjun","Simran","Aman","Zoya"],
    "City": ["Delhi","Mumbai","Delhi","Pune","Mumbai","Delhi","Pune","Delhi","Mumbai","Pune","Delhi","Mumbai"],
    "Category": ["Electronics","Clothing","Electronics","Furniture","Clothing","Electronics","Furniture","Clothing","Electronics","Furniture","Clothing","Electronics"],
    "Amount": [25000,3000,18000,22000,4500,27000,15000,3500,20000,24000,3200,26000],
    "Quantity": [1,2,1,1,3,1,2,2,1,1,2,1],
    "PaymentMode": ["UPI","Card","UPI","Cash","Card","UPI","Cash","UPI","Card","Cash","Card","UPI"]
})

sales_df
```

```python
# Average Amount Per Category
pd.pivot_table(sales_df,values="Amount",index="Category",aggfunc="mean")
```

```python
# Total Amount Per Category
pd.pivot_table(sales_df,values="Amount",index="Category",aggfunc="sum")
```

```python
# Multiple Aggregation
pd.pivot_table(sales_df,values="Amount",index="Category",aggfunc=["mean","sum","max"])
```

```python
# Multiple Column Pivot
# Avg Amount + Avg Quantity
pd.pivot_table(sales_df,values=["Amount","Quantity"],index="Category",aggfunc="mean")
```

```python
# Matrix Pivot (Excel Style)
# Category vs City Avg Amount
pd.pivot_table(sales_df,values="Amount",index="Category",columns="City",aggfunc="mean",fill_value=0)
```

```python
# Revenue by Category vs Payment Mode
pd.pivot_table(sales_df,values="Amount",index="Category",columns="PaymentMode",fill_value=0)
```
