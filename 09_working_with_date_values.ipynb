9. working with date values

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "Name": ["Aman","Riya","Kunal","Sneha"],
    "DOJ": ['2024-01-01', '2024-01-15', '2024-03-28', '2024-03-03'],
    "Marks": [78,85,90,70]
})
```

```python
df["DOJ"].dtype
```

```python
df["DOJ"] = pd.to_datetime(df["DOJ"])
df["DOJ"].dtype
```

```python
# Handling different date formats
df["DOB"] = ["21-10-2002", '22-11-2000','28-09-2003','02-12-2005']
df["DOB"] = pd.to_datetime(df["DOB"],format="%d-%m-%Y")
df
```

```python
# Extract parts of date (Year, Month, Dayâ€¦)
df["Year"] = df["DOJ"].dt.year
df["Month"] = df["DOJ"].dt.month
df["Day"] = df["DOJ"].dt.day
df
```

```python
# df["DOJ"].dt.day_name()
# df["DOJ"].dt.month_name()
```

```python
# Date-based filtering (VERY IMPORTANT)
# Filter data after a specific date
print(df[df["DOJ"] > "2024-01-20"])
```

```python
# Filter between two dates
df[(df["DOB"]>= "2002-12-29") & (df["DOB"]<="2004-10-30")]
```

```python
# Current Date & Time
pd.Timestamp.now()
```

```python
pd.Timestamp.today().date() #only date
```

```python
# Difference between two dates (Time Delta)
df["Date_of_prize"] =  pd.Timestamp("2025-12-30") - df["DOJ"]
df
```

```python
# Sorting by Date (common use)
df.sort_values(by="DOJ",inplace=True)
df.reset_index(drop=True,inplace=True)
df
```
