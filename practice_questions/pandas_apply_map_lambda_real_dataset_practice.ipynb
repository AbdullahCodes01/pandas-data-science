```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "EmpID": [101,102,103,104,105,106,107,108,109,110],
    "Name": ["Arman","Riya","Kunal","Sneha","Vikram","Pooja","Rahul","Neha","Arjun","Simran"],
    "Department": ["IT","HR","IT","Finance","HR","IT","Finance","HR","IT","Finance"],
    "Salary": [60000,45000,72000,68000,50000,58000,75000,52000,40000,69000],
    "Experience": [2,1,5,4,2,3,7,2,1,6],
    "Rating": [4.5,3.8,4.9,4.2,3.5,4.0,4.8,3.9,3.2,4.6]
})

df
```

```python
# Q1Create column Salary_Level
# salary ≥ 70000 → High
# 50000 – 69999 → Medium
# < 50000 → Low

df["Salary_Level"] = df["Salary"].apply(lambda x: "High" if x>=70000 else "low" if x<50000 else "Medium")
df
```

```python
# Create Bonus column
#  Bonus = 12% Salary
df["Bonus"] = df["Salary"].apply(lambda x: x*0.12)
df
```

```python
# Create Experience_Level
# Experience ≥ 5 → Senior
# 2 – 4 → Mid
# < 2 → Junior

df["Experience_Level"] = df["Experience"].apply(lambda x: "Senior" if x>=5 else "Junior" if x<2 else "Mid")
df
```

```python
# Create Performance column

# Rating ≥ 4.5 → Excellent
# 4 – 4.49 → Good
# else → Average

df["Performance"] = df["Rating"].apply(lambda x: "Excellent" if x>=4.5 else "Average" if x<4 else "Good")
df
```

```python
# IT employees → Salary increase 10%
df["New_Salary"] = df["Salary"].apply(lambda x: x*1.10)
df
```

```python
# Create Tax column

# IT → 10% Salary
# HR → 8% Salary
# Finance → 12% Salary
df["Tax"] = df.apply(lambda row: row["Salary"]*0.10 if row["Department"] == "IT" else row["Salary"]*0.08 if row["Department"] == "HR" else row["Salary"]*0.12,axis=1)
df
```

```python
# Average salary per department
avg_salary = df.groupby("Department")["Salary"].mean().reset_index()
avg_salary
```

```python
# Top 3 highest paid employees
df.sort_values("Salary",ascending=False).head(3)[
    ["EmpID", "Department", "Salary"]
]
```

```python
# Employees whose salary > department average
df[df["Salary"] > df.groupby("Department")["Salary"].transform("mean")][
    ["Name","Salary","Department"]
]
```

```python
# Create Risk column
# Rating < 3.5 → High Risk
# 3.5 – 4 → Medium Risk
# > 4 → Low Risk
df["Risk"] = df["Rating"].apply(lambda x: "High Risk" if x<3.5 else "Low risk" if x>4 else "Medium Risk")
df
```

```python
# Find:
# Department with highest average rating
# df.groupby("Department")["Rating"].mean().sort_values(ascending=False)
df.groupby("Department")["Rating"].mean().idxmax()
```

```python
# Create Promotion column
# Eligible if:
# ✔ Experience ≥ 4
# ✔ Rating ≥ 4.5
# ✔ Salary < 70000
# Else → Not Eligible

df["Promotion"] = np.where(
    (df["Experience"]>=4) & (df["Rating"]>=4.5) & (df["Salary"]<70000),
    "Eligible","Not Eligible"
)
    
df
# alternative 
# df["Promotion"] = df.apply(
#     lambda row: "Eligible"
#     if (row["Experience"] >= 4 and row["Rating"] >= 4.5 and row["Salary"] < 70000)
#     else "Not Eligible",
#     axis=1)


# df["Promotion"] = "Not Eligible"

# df.loc[
#     (df["Experience"] >= 4) &
#     (df["Rating"] >= 4.5) &
#     (df["Salary"] < 70000),
#     "Promotion"
# ] = "Eligible"
```
