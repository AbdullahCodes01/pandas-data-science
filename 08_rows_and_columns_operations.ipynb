8. Rows & Columns Operations in Pandas

```python
import pandas as pd
import numpy as np
```

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aman","Riya","Kunal","Sneha"],
    "Age": [20,21,22,23],
    "Marks": [78,85,90,70]
})

df
```

PART 1: ADD OPERATIONS

1. Add a NEW COLUMN (MOST COMMON)

```python
# (a) Add a NEW COLUMN (MOST COMMON)
df["city"] = ["delhi","noida","mumbai","delhi"]
df
```

```python
# (B) Column from calculation
df["Result"] = df["Marks"] >= 80
df
```

```python
# (C) Column with default value
df["passed"] = "Yes"
df
```

```python
# (D) Column using condition
df["Grade"] = "C"
df.loc[df["Marks"]>85,"Grade"] = "A"
df.loc[(df["Marks"]>=75) & (df["Marks"]<85),"Grade"] = "B"
df
```

2. Add a NEW ROW

```python
# (A) Using .loc 
df.loc[len(df)] = ["arun",22,88,"pune",True,"Yes","A"]
```

```python
df
```

```python
# (B) Using concat() (Professional way)
new_row = pd.DataFrame([{
    "Name": "Raghav",
    "Age": 24,
    "Marks": 77,
    "city": "Pune",
    "Result": False,
    "passed": "yes",
    "Grade": "B"
}])

df = pd.concat([df,new_row],ignore_index=True)
df


```

PART 2: UPDATE OPERATIONS

```python
# Update a COLUMN (entire column)
df["Age"] = df["Age"] + 1
df
```

```python
# Update specific rows using condition ( VERY IMPORTANT)
# Increase marks where Marks < 75
df.loc[df["Marks"]<75,"Marks"] = 75
```

```python
# Add bonus marks to Delhi students
df.loc[df["city"]=="delhi","Marks"] += 1
```

```python
# Update single cell
# using loc
df.loc[0,"Marks"] = 90
```

```python
# Using iloc
df.iloc[1,3] = "87"
```

```python
df
```

PART 3: DELETE OPERATIONS

```python
# Delete a COLUMN
df.drop("Result",axis=1,inplace=True)
```

```python
df
```

```python
# Delete multiple columns
df.drop(["passed","Grade"],axis=1,inplace=True)
df
```

```python
# Delete a ROW (by index)
df.drop(1,axis=0,inplace=True)
```

```python
df
```

```python
# Delete rows using condition
df.drop(df[df["Marks"]<80].index,inplace=True)
```

```python
df
```

```python
# Delete rows with missing values
df.dropna(inplace=True)
```
