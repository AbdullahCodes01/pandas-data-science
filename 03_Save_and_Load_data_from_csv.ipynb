3. Save and Load data from csv

```python
import pandas as pd
import numpy as np
```

```python
data = {
     "Name": ["Tushar","vaibhav","deepika","Abdullah"],
     "roll_no": [1,2,3,4],
     "Age": [20,20,19,21],
     "marks": [80,85,73,93]
 }

df = pd.DataFrame(data)
print(df)
```

```python
df.to_csv('test_data.csv',index=False) #save file - export dataframe
```

```python
load_df = pd.read_csv("test_data.csv") #load file - import dataframe
load_df
```

```python
load_df.head(2) # first 2 rows
```

```python
load_df.tail(2) #last 2 rows
```

```python
df1 = pd.read_csv("test_data.csv",usecols=["Name","marks"]) # to read selected columns
df1 
```

```python
# df2 = pd.read_csv("test_data.csv",header=None)
df2 = pd.read_csv("test_data.csv", header=None, names=["A","B","C"])
df2
```

```python
df.to_csv("output.csv",index=False) #without index 
#df.to_csv("output.csv",sep = ";") custom seprator
```

```python
df.to_csv("selected_column.csv",columns=["Name","marks"]) # to save selected column
```

```python
# Real-life example
data1 = {
    "name": ["Arman","Akansha","pritish"],
    "Marks": [75,88,65],
    "city": ["delhi","delhi","up"]
}

df3 = pd.DataFrame(data1)
# Save
df3.to_csv("students.csv",index=False)

#Load again
new_df = pd.read_csv("students.csv",usecols=["name","Marks"])
print(new_df)
```

```python
# Important parameters 

# For reading:

# sep

# usecols

# header

# names

# nrows

# For saving:

# index

# sep

# columns
```
