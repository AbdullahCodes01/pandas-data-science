```python
import pandas as pd
```

1. Value_counts( )

```python
# value_counts() — Frequency Count
df = pd.DataFrame({
   "department": ["IT","HR","IT","Finance","HR","IT","Finance"] 
})
print(df.value_counts())
```

```python
# Percentage Distribution
df["department"].value_counts(normalize=True) * 100
```

```python
# Top N Values
df["department"].value_counts().head(2)
```

```python
df["department"].value_counts(dropna=False)
# including missing values
```

```python
age = pd.Series([21,22,23,25,30,35,40,45])
age.value_counts(bins=3)
# Bins (Numeric Data)
```

2. crosstab() — Two Column Frequency Relationship

```python
sales_df = pd.DataFrame({
    "OrderID": [1001,1002,1003,1004,1005,1006,1007,1008,1009,1010,1011,1012],
    "Customer": ["Arman","Riya","Kunal","Sneha","Vikram","Pooja","Rahul","Neha","Arjun","Simran","Aman","Zoya"],
    "City": ["Delhi","Mumbai","Delhi","Pune","Mumbai","Delhi","Pune","Delhi","Mumbai","Pune","Delhi","Mumbai"],
    "Category": ["Electronics","Clothing","Electronics","Furniture","Clothing","Electronics","Furniture","Clothing","Electronics","Furniture","Clothing","Electronics"],
    "Amount": [25000,3000,18000,22000,4500,27000,15000,3500,20000,24000,3200,26000],
    "Quantity": [1,2,1,1,3,1,2,2,1,1,2,1],
    "PaymentMode": ["UPI","Card","UPI","Cash","Card","UPI","Cash","UPI","Card","Cash","Card","UPI"]
})
sales_df
```

```python
pd.crosstab(sales_df["PaymentMode"],sales_df["City"])
```

```python
# Normalize Crosstab (Percentage)
pd.crosstab(sales_df["City"],sales_df["Category"],normalize=True) * 100
```

```python
# Add Totals
pd.crosstab(sales_df["City"],sales_df["Category"],margins=True)
```

```python
# Crosstab With Aggregation ( Powerful)
pd.crosstab(sales_df["City"],sales_df['Category'],values = sales_df["Amount"],aggfunc="mean").fillna(0)
# city and category wise avg selling ammount
```
